Create a program that converts number grades to letter grades.

### Console:
```powershell

Enter numerical grade: 90

Continue? (y/n): y

Enter numerical grade: 88 
Letter grade: A

Continue? (y/n): y

Enter numerical grade: 80

Enter numerical grade: 67 
Letter grade: C

Continue? (y/n): y

Enter numerical grade: 59 
Letter grade: F

Continue? (y/n): n

```

### Specifications:
- Assume that the user will enter valid integers for the grades.
- The program should continue only if the user enters “y” or “Y” to continue.
- The grading criteria is as follows:

| Letter |	Range |
|---|-------|
| A |	88-100 |
| B |	80-87 |
| C |	67-79 |
| D |	60-66 |
| F |	<60 |

````python
'''While loop'''
print("Letter Grade Converter\n")

sContinue="Y"
#Whiloe loop
while sContinue == "Y" :#or sContinue == "y":
    iGrade = int(input("Enter numerical grade: "))
    #If statement
    if iGrade <= 100 and iGrade >=88:
        print("Letter grade: A\n")
    elif iGrade <= 87 and iGrade >=80:
        print("Letter grade: B\n")
    elif iGrade <= 79 and iGrade >=67:
        print("Letter grade: C\n")
    elif iGrade <= 66 and iGrade >=60:
        print("Letter grade: D\n")
    elif iGrade <= 60:
        print("Letter grade: F\n")
    sContinue =input("Continue? (y/n)").upper()
    #sContinue.upper()
print("\nBye!")
````

## 3.2 - Tip Calculator
Create a program that calculates three options for an appropriate tip to leave after a meal at a restaurant.

### Console
```powershell
Tip Calculator

Cost of meal: 52.31 

15%
Tip amount:	7.85
Total amount: 60.16

20%
Tip amount:	10.46
Total amount: 62.77

25%
Tip amount:	13.08
Total amount: 65.39
```

### Specifications:
- The program should calculate and display the cost of tipping at 15%, 20%, or 25%.
- Assume the user will enter valid data.
- The program should round results to a maximum of two decimal places.



````python
'''Tip calculatopr'''
print("Tip calcultor\n")
#Get cost of meal from user
fCostMeal = float(input("Cost of meal: "))
#calculation
for i in range(15,30,5):
    print(i,"%")
    fTip = fCostMeal * i/100
    print("Tip amount: ",round(fTip,2))
    fTotal = fCostMeal + fTip
    print("Total amount: ", round(fTotal,2))
````

```powershell

Enter number of cents (0-99): 55

Quarters: 2

Bye!
```

### Specifications
- The program should display the minimum number of quarters, dimes, nickels, and pennies that one needs to make up the specified number of cents.
- Assume that the user will enter a valid integer for the number of cents.
- The program should continue only if the user enters “y” or “Y” to continue.


````python
'''Change calculator'''
sTemp = "y"
#start while loop
while sTemp == "y":
    #Define var
    iQuarters = 0
    iDimes = 0
    iNickels = 0
    iPennies = 0
    
    #user input, get total cents
    iCents = int(input("Enter number of cents (0-99): "))
    
    #calculations
    #Quarter
    iTemp = iCents % 25
    iQuarters = round((iCents - iTemp)/25)
    iCents -= iTemp
    
    #Dimes
    iTemp = iCents % 10
    iDimes = round((iCents - iTemp)/10)
    iCents -= iTemp
    
    #Nickels
    iTemp = iCents % 5
    iNickels = round((iCents - iTemp)/5)
    iCents -= iTemp
    
    #Pennies
    iPennies = iCents
    
    #Display
    print()
    print("Quarters:",iQuarters)
    print("Dimes:",iDimes)
    print("Nickels",iNickels)
    print("Pennies:",iPennies)
    print()
    
    #Contenue?
    sTemp = input("Contenue? (y/n): ")
    sTemp.lower
    
````
